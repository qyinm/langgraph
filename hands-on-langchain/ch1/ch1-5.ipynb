## 명령형 구성
평범하게 구성 요서를 함수와 클래스로 결합하는 행위.  
프롬프트, 모델, 출력파서를 결합하는 예시

```python
from langchain_openai.chat_models import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.runnables import chain

# 구성 요소
template = ChatPromptTemplate.from_messages([
    ("system", "당신은 친절한 어시스턴트입니다."),
    ("human", "{question}")
])

model = ChatOpenAI(model="gpt-4.1-mini")

# 함수로 결합
# 데코레이터 @chain을 추가해 작성한 함수에 Runnable 인터페이스 추가
@chain
def chatbot(values):
    prompt = template.invoke(values)
    return model.invoke(prompt)

chatbot.invoke({"question": "거대 언어 모델은 어디서 제공하나요?"})
```

명령형 구성을 사용한 스트리밍 호출 예시

```python
@chain
def chatbot(values):
    prompt = template.invoke(values)
    for token in model.stream(prompt):
        yield token

for part in chatbot.stream({
    'question': '거대 언어 모델은 어디서 제공하나요?'
}):
    print(part)
```

명령형 구성에서 비동기 실행  
JS는 비동기만 실행 지원해서. 비동기 실행방식은 파이썬에서만 지원됨.

```python
@chain
async def chatbot(values):
    prompt = await template.ainvoke(values)
    return await model.ainvoke(prompt)

await chatbot.ainvoke({
    'question': '거대 언어 모델은 어디서 제공하나요?'
})
```

## 선언형 구성
LCEL(LangChain Expression Language)은 선언형 언어를 이용해 랭체인 요소 구성.
랭체인이 LCEL로 최적화되게 컴파일함. 자동 병렬화, 스트리밍, 추정 및 비동기 지원 수행

```python
from langchain_openai.chat_models import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate

template = ChatPromptTemplate.from_messages([
    ("system", "당신은 친절한 어시스턴트입니다."),
    ("human", "{question}")
])

model = ChatOpenAI(model="gpt-4.1-mini")

chatbot = template | model

chatbot.invoke({"question": "거대 언어 모델은 어디서 제공하나요?"})
```

```python
chatbot = template | model

for part in chatbot.stream({
    'question': '거대 언어 모델은 어디서 제공하나요?'
}):
    print(part)
```
